```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
plt.style.use('classic')
```

```python
x = np.linspace(0, 10, 100)
```

```python
fig = plt.figure()

plt.plot(x, np.sin(x), '-')
plt.plot(x, np.cos(x), '--')

plt.show()
```

```python
fig.savefig('my_figure.png')
```

```python
plt.figure()

plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
```

```python
# better way
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```
