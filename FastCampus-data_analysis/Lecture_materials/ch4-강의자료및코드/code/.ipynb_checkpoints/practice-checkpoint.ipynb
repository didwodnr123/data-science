## How to deal with Categorical values

- Gender = Male:1, Female:0
- Season = Spring:[1,0,0,0], Summer:[0,1,0,0] ...

```python
data['gender'].nunique()
```

```python
# get rid of overlapped columns, 
# such as gender_male & gender_female
# we need only one column.
pd.get_dummies(data, columns=['gender'], drop_first=True)
```

```python
data['gender'].dtype == 'O'
```

```python
col_list = []
```

```python
data.columns
```

```python
# get all 'Object' columns
for i in data.columns:
    if data[i].dtype == 'O':
        col_list.append(i)
```

```python
for i in col_list:
    print(i, data[i].nunique())
```

```python
scaled_data = pd.DataFrame(minmax.transform(data), columns=data.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = scaled_data.drop('Churn_Yes', axis=1)
y = scaled_data['Churn_Yes']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train, y_train)
```

```python
pred = knn.predict(X_test)
```

```python
pd.DataFrame({'actual_value':y_test, 'pred_value':pred}).head(10)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test, pred)
```

```python
confusion_matrix(y_test, pred)
```

```python
error_list = []
```

```python
for i in range(1, 101):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_list.append(accuracy_score(y_test, pred))
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=range(1,101), y=error_list, marker='o', markersize=10, markerfacecolor='red')
```

```python
error_list.index(max(error_list))
```

```python
np.array(error_list).argmax()
```
