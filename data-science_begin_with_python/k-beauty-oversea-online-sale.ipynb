# 데이터 로드하기

```python
df_raw = pd.read_csv('../data/국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20201220190028.csv', 
                     encoding='cp949')
df_raw.shape
```

```python
df_raw[df_raw['국가(대륙)별'] == '미국']
```

# 분석과 시각화를 위한 tidy data 만들기

```python
df_raw.columns
```

```python
df_raw.values
```

```python
df = df_raw.melt(id_vars=['국가(대륙)별', '상품군별', '판매유형별'], 
                 var_name='기간', 
                 value_name='백만원')
df.head()
```

```python
df.info()
```

# 데이터 전처리
## 기간에서 연도를 분리하기

```python
int('2019 4/4'.split()[0])
```

```python
df['연도'] = df['기간'].map(lambda x : int(x.split()[0]))
df.head()
```

```python
'2019 1/4'.split()[1].split('/')[0]
```

```python
df['분기'] = df['기간'].map(lambda x : int(x.split()[1].split('/')[0]))
df.head()
```

## 금액을 수치데이터로 표현하기 위해 테이터 타입 변경하기

```python
# NaN == float이기 때문에 전부 float 데이터로 변경시켜준다.
df['백만원'] = df['백만원'].replace('-', np.nan).astype(float)
df.head()
```

```python
df.info()
# memory check
```

```python
# 합계 데이터는 따로 구할 수 있기 때문에 전체 데이터에서 제거한다.
df = df[(df['국가(대륙)별'] != "합계") & (df['상품군별'] != '합계')].copy()
df.info()
```

```python
# null value check
df.isnull().sum()
```

# K-Beauty 시각화
## 전체 상품군 판매액

```python
# 결측치를 제거해야한다.
df_total = df[df['판매유형별'] == '계'].copy()
df_total.head()
```

```python
# 연도, 판매액으로 lineplot 그리기
sns.lineplot(data=df_total, x='연도', y='백만원')
```

```python
sns.lineplot(data=df_total, x='연도', y='백만원', hue='상품군별')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.relplot(data=df_total, x='연도', y='백만원', 
            hue="상품군별", kind='line', col='상품군별', col_wrap=4)
```

```python
# 화장품과 의류 및 패션이 그래프에 영향을 크게 미치기 때문에 제거.
df_sub = df_total[~df_total['상품군별'].isin(['화장품', '의류 및 패션 관련상품'])].copy()
```

```python
sns.relplot(data=df_sub, x='연도', y='백만원', 
            hue='상품군별', col='상품군별', col_wrap=4, kind='line')
```
